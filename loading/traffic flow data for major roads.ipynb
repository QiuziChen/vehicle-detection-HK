```python
import requests
import xmltodict
```

```python
raw_url = 'https://resource.data.one.gov.hk/td/traffic-detectors/rawSpeedVol-all.xml'
pro_url = 'https://resource.data.one.gov.hk/td/traffic-detectors/irnAvgSpeed-all.xml'
```

```python
def get_dict(url):
    res = requests.get(url)
    data = xmltodict.parse(res.content)
    return data
```

```python
raw = get_dict(raw_url)
pro = get_dict(pro_url)
```
